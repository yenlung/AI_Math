# 畫一張圖

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

# 接力畫一張完整的圖

```python

plt.figure(figsize=(6,6))
plt.xkcd()#設定圖的大小是(6, 6)
#設定一個風格

plt.plot(x, y1,label='sin(x)')#y1的名字是"sin"
plt.plot(x, y2, label='cos(x)')#y2的名字是"cos"
plt.legend('lower left')#將圖標放在左下角
plt.xlabel("X Axis")
plt.ylabel('Y Axis')
#X軸的名字是"X Axis"
#Y軸的名字是"Y Axis"


plt.xlim(0,10)
plt.ylim(-1, 1)
#設定X軸的範圍
#設定Y軸的範圍

plt.xticks(np.arange(0,10,1))#X軸分成10個部分
plt.yticks(np.arange(-1,1,0.5))#Y軸分成4個部分

plt.title('Sin and Cos')#圖的標題是"Sin and Cos"
```

# Variable Length Function Arguments

```python
def f(x):
    print(x)
```

```python
f("Hello")
```

```python
f("Hello", "World")
```

```python
def ARGV(*argv):
    for i in argv:
        print(i)
```

```python
ARGV("Hello", "World")
```

```python
ARGV("Hello", "World", "I", "am", "hungry")
```

```python
def ARGV1(arg1, *argv):
    print(arg1)
    
    print("")
    for i in argv:
        print(i)
```

```python
ARGV1("Hello", "World", "I", "am", "hungry")
```

```python
def ARGV2(*argv, arg1):
    print(arg1)
    
    print("")
    for i in argv:
        print(i)
```

```python
ARGV2("Hello", "World", "I", "am", "hungry", arg1 = "YA")
```

# Variable Length Keyword Argument

```python
def KWARG(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} == {value}")
```

```python
KWARG(first = "N", second = "C", third = "T", fourth = "S")
```

```python
def Four(arg1, arg2, arg3, arg4):
    print("arg1 is ", arg1)
    print("arg2 is ", arg2)
    print("arg3 is ", arg3)
    print("arg4 is ", arg4)
```

```python
args = ("N", "T", "C", "S")
```

```python
Four(*args)
```

```python
kwargs = {"arg2" : "T", "arg1" : "N", "arg4" : "S" ,"arg3" : "C",}
```

```python
Four(**kwargs)
```
